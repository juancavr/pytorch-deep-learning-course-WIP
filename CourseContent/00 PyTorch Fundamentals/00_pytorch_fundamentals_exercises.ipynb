
### 6. Create two random tensors of shape `(2, 3)` and send them both to the GPU (you'll need access to a GPU for this). Set `torch.manual_seed(1234)` when creating the tensors (this doesn't have to be the GPU random seed). The output should be something like:

```
Device: cuda
(tensor([[0.0290, 0.4019, 0.2598],
         [0.3666, 0.0583, 0.7006]], device='cuda:0'),
 tensor([[0.0518, 0.4681, 0.6738],
         [0.3315, 0.7837, 0.5631]], device='cuda:0'))
```

````python
# Set random seed
torch.manual_seed(1234)
tensor_rnd_5 = torch.rand(2,3)
# Every time you reset the seed, forces the pseudorandom number generator to start from the begining,
# thats why if we do it here we will get the same results in both cases.
# Usually this type of pseudorandom number generators are continous probability distributions
# and the seed indicates from where to start picking numbers.
#torch.manual_seed(1234)
tensor_rnd_6 = torch.rand(2,3)

# Check for access to GPU
device = "cuda" if torch.cuda.is_available() else "cpu"

# Create two random tensors on GPU

tensor_on_gpu_5 = tensor_rnd_5.to(device)
tensor_on_gpu_6 = tensor_rnd_6.to(device)

device,tensor_on_gpu_5,tensor_on_gpu_6
````


### 7. Perform a matrix multiplication on the tensors you created in 6 (again, you may have to adjust the shapes of one of the tensors).

### 8. Find the maximum and minimum values of the output of 7.

````python
# Find max
max = result.max()
# Find min
min = result.min()

print(f'Max: {max} | Min: {min}')
````

### 9. Find the maximum and minimum index values of the output of 7.

````python
# Find arg max
argmax = result.argmax()
# Find arg min
argmin = result.argmin()

print(f'Index of Max: {argmax} | Index of Min: {argmin}')
````
